```python
'''
1. Write a program to add two lists index-wise. Create a new list that contains the 0th index item from both the list, then the 1st index item, and so on till the last element. any leftover items will get added at the end of the new list.

	Given List:

	list1 = ["M", "na", "i", "Ra"]
	list2 = ["y", "me", "s", "hul"]

	Output:

	[['M','y'], ['na', me'], ['i', 's'], ['Ra', 'hul']]
'''

list1 = ["M", "na", "i", "Ra"]
list2 = ["y", "me", "s", "hul"]

# using empty list and for loop
re  = []
for i in range(len(list1)):
    for j in range(len(list2)):
        if i == j:          # both index checking
            re.append([list1[i],list2[j]])
print(re)

# above list comprehension
re = [[list1[i],list2[j]] for i in range(len(list1)) for j in range(len(list2))  if i==j ]
print(re)

# using zip
re = [[a,b] for a,b in zip(list1,list2)]
print(re)
```

```python
'''3. Suppose you are given a list of candy and another list of same size representing no of items of respective candy.

	i.e -

	candy_list = ['Jelly Belly','Kit Kat','Double Bubble','Milky Way','Three Musketeers']
	no_of_items = [10,20,34,74,32]
	Write a program to show no. of items of each candy type.

	Output:

	Jelly Belly-10
	Kit Kat-20
	Double Bubble-34
	Milky Way-74
	Three Musketeers-32 '''

candy_list = ['Jelly Belly','Kit Kat','Double Bubble','Milky Way','Three Musketeers']
no_of_items = [10,20,34,74,32]


for a,b in zip(candy_list,no_of_items):
    print(f"{a}-{b}")
```

```python
'''
4. Running Sum on list Write a program to print a list after performing running sum on it.

	i.e:

	Input:

	list1 = [1,2,3,4,5,6]
	Output:

	[1,3,6,10,15,21]
'''

list2 = []
list1 = [1,2,3,4,5,6]
total = 0
for i in list1:
    total += i
    list2.append(total)
print(list2)

# in place
list3 = [1,2,3,4,5,6]
for i in range(1,len(list3)):
    list3[i] = list3[i] + list3[i-1]
print(list3)
```

```python
'''
6. Find list of common unique items from two list. and show in increasing order

	Input

	num1 = [23,45,67,78,89,34]
	num2 = [34,89,55,56,39,67]
	Output:

	[34, 67, 89]
'''
num1 = [23,45,67,78,89,34]
num2 = [34,89,55,56,39,67]

l = []
l2 =[]
for x in num1:
    if x in num2:
        l.append(x)
    
print(sorted(l))


# ==================================
l4 =[x for x in num1 if x in num2]
print(sorted(l4))

# =========================================
l6  = list(filter(lambda z: z in num2,num1 ))
print(sorted(l6))

```

```python
'''
7. Sort a list of alphanumeric strings based on product value of numeric character in it. If in any string there is no numeric character take it's product value as 1.

	Input:

	['1ac21', '23fg', '456', '098d','1','kls']
	Output:

	['456', '23fg', '1ac21', '1', 'kls', '098d']

    You have a list of strings that contain both letters and numbers mixed together. You need to sort this list based on the product of all numeric digits found in each string.
    '1ac21'  → digits are: 1, 2, 1
    '1ac21'  → 1+2+1 = 4
    
    '''

l =['1ac21', '23fg', '456', '098d','1','kls']

for x in l:
    if x.isalnum():
        print(x)
        for y in x:
            total = 0
            if y.isdigit():
                total += int(y)
        print(total)
```

```python
'''
8. Write a program that can find the max number of each row of a matrix
	Example:

	Input:

	[[11,22,33],[44,55,66],[77,88,99]]
	Output:

	[3,6,9]
'''
m = [[11,22,33],[44,55,66],[77,88,99]]
re = [max(row) for row in m]
print(re)

```
