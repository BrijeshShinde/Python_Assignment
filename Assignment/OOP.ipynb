```python
'''
4. Write a Python program to create a class that represents a shape. Include methods to calculate its area and perimeter. Implement subclasses for different shapes like circle, triangle, and square.'''

from abc import ABC,abstractmethod
from math import pi

class Shape(ABC):
    
    @abstractmethod
    def area(self):pass
    
    @abstractmethod
    def perimeter(self):pass

class Circle(Shape):
    def __init__(self,r):
        self.r =r

    def area(self):
        area = pi* self.r * self.r
        return f'Area of circle is {area:.2f}'
    
    def perimeter(self):
        perimeter = 2 * pi * self.r
        return f"Perimeter of circle is {perimeter:.2f}"

class Triangle(Shape):
    def __init__(self,b,h,c):
        self.b = b
        self.h = h
        self.c = c
   
    def area(self):
        area = 0.5 * self.b * self.h
        return f"Area of Triangle is {area}"

    def perimeter(self):
        perimeter = self.b + self.h + self.c
        return f"Perimeter of Triangle is {perimeter}"

class Squre(Shape):
    def __init__(self,s):
        self.s = s
    
    def area(self):
        area = self.s * self.s
        return f"Area of squre is {area}"
    
    def perimeter(self):
        perimeter = 4 * self.s
        return f"Perimeter of squre is {perimeter}"

        
c1 = Circle(2)
print(c1.area())
print(c1.perimeter())

t1 =Triangle(24,12,36)
print(t1.area())
print(t1.perimeter())

s1 =Squre(2)
print(s1.area())
print(s1.perimeter())
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

print(is_prime(7))
print(is_prime(10))
```
