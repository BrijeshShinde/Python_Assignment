```python
'''
1. Write a Python program to handle a ZeroDivisionError exception when dividing a number by zero.
'''
def division(divident,divisor):
        try:
            print(divident/divisor)
        except ZeroDivisionError:
            print("Number Divied by Zero Not Possible")

div = 5
dsor = 0
print(division(div,dsor))
```

```python
'''
2. Write a Python program that prompts the user to input an integer and raises a ValueError exception if the input is not a valid integer
'''
try:
    n = int(input("Enter int number:"))
except ValueError as e:
    print("Enter integer number only")
else:
    print(f"You have entered {n}")
```

```python
'''
4. Write a Python program that prompts the user to input two numbers and raises a TypeError exception if the inputs are not numerical.
'''
def nums():
    try :
        num1 = input("Enter num1: ")
        num2 = input("Enter num2: ")
        x = num1 - num2
    except TypeError :
        print("Type Error : Enter Proper integer number ")
    else:
        print(f" {x} = {num1}+{num2}")

nums()
```

```python
'''
5. Write a Python program that opens a file and handles a PermissionError exception if there is a permission issue.
'''
import os
os.chmod('sample.txt',0o444)

try:
    with open("sample.txt",'r') as f:
        data = f.read()
        print(data)

except PermissionError:
    print("You dont have permission to access this file")
except FileNotFoundError:
    print("File not found")


# testing error
try:
    with open('sample.txt','w') as f:
        f.write('Hello')
except PermissionError:
    print("Acccess denied")
```

```python
'''
6. Write a Python program that executes an operation on a list and handles an IndexError exception if the index is out of range
'''
l = [[11,22],[33,44]]

try:
    for x in l:
        for y in x:
            print(l[y])
except IndexError :
    print("Index out of range")
```

```python
'''
7. Write a Python program that executes division and handles an ArithmeticError exception if there is an arithmetic error.
'''
def cal_avg(marks):
    try :
        total =sum(marks)
        avg = total/len(marks)
        return avg
    except ArithmeticError as e:
        print(f"Error : {e} ")

print(cal_avg([80,20,50]))
print(cal_avg([]))
```
