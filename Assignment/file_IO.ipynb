```python
'''
11. Write a Python program to write a list to a file.
'''
def writel(l:list):
    with open('fileio1.txt','a+') as f:
        # get list of item convert them into str then write file with \n
        for y in l:
            s = str(y)
            f.writelines('\n'+s)

        # read file
        f.seek(0)
        data = f.readlines()
        for x in data:
            print(x,end="")


l = [11,'HI',True]
writel(l)
```

```python
'''
12. Write a Python program to copy the contents of a file to another file .
'''
def copyf():
    with open("fileio1.txt",'r') as f:
        data = f.read()

        with open("fileio1(copy).txt",'w+') as g:
            g.write(data)
            g.seek(0)
            file = g.readlines()
            for x in file:
                print(x,end="")
copyf()
```

```python
'''
14. Write a Python program to assess(access) if a file is closed or not.
'''
def check(file):
    with open(file,'r') as f:
        print(f'file is opened inside with : {f.closed}')
    print(f'file is closed outside with: {f.closed}')

check('fileio1.txt')
```

```python
'''
15.Write a Python program that takes a text file as input and returns the number of words of a given text file.
'''
def no_wordsf(file):
    with open(file,'r') as f:
        # readlines into list
        data = f.readlines()
        # maintain counter
        cnt =0
        # run for loop for one line
        for x in data:
                # spliting x[0] by space which count words and stores in 'words'
                words  = x.split()
                # adding len(words) to counter
                cnt+= len(words)
        return f"no of words are {cnt} "
no_wordsf('fileio1.txt')
```
