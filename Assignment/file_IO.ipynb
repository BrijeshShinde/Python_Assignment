```python
'''
2. Write a Python program to read first n lines of a file.

'''
n = int(input("enter no. of lines to read"))
with open('fileio1.txt','r') as f:
    for i in range(n):
        data = f.readline()
        print(data,end='')
```

```python
'''
3. Write a Python program to append text to a file and display the text.
'''
def addline(s:str):
    with open('fileio1.txt','a+') as f:
        f.writelines(f'{s}')
        print("Your Text appended sucessfully ....")
        f.seek(0)
        f.tell()
        data = f.read()
        print(data,end='')

s = 'Its alarming so, AI developers should add limits for AGI from doing illegal work.'
s2 = 'or should add guidelines for AI while developing'
addline(s2)
```
