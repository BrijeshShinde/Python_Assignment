```python
'''
8. Write a Python function that checks whether a passed string is a palindrome or not.

Note: A palindrome is a word, phrase, or sequence that reads the same backward as forward, e.g., madam or nurses run
'''

s = "Madam"

def palindrome(s):
    if s.lower() == s[::-1].lower():
        print("Palindrom")
    else:
        print("Not palindrom")

palindrome(s)


def palindrom2(s):
    left, right = 0,len(s)-1
    while left < right:
        if s[left].lower() != s[right].lower():
            return False
        left +=1
        right -=1
    return True

s = 'madam'
palindrom2(s)
```

```python
'''
12. Write a Python program to create a function that takes one argument, and that argument will be multiplied with an unknown given number.

	Sample Output:
	Double the number of 15 = 30
	Triple the number of 15 = 45
	Quadruple the number of 15 = 60
	Quintuple the number 15 = 75



The requirement is to create a function that returns another function (a closure/higher-order function).
Step by step:

Create a function that takes one argument (the multiplier — like 2, 3, 4, 5)
That function returns a new function which multiplies any number by that multiplier
Then use it like:

Pass 2 → get a "doubler" function → call it with 15 → gives 30
Pass 3 → get a "tripler" function → call it with 15 → gives 45
Pass 4 → get a "quadrupler" function → call it with 15 → gives 60
Pass 5 → get a "quintupler" function → call it with 15 → gives 75
'''


def multiplier(n):
    def mul(x):
        return n * x
    return mul


double = multiplier(2)
print(double(15))

triple = multiplier(3)
print(triple(15))

quadruple = multiplier(4)
print(quadruple(15))

Quintuple = multiplier(5)
print(Quintuple(15))
```

```python
# aggregation(losse coupling)
class Customer:
    def __init__(self,name,address):
        self.name = name
        self.address=address
    def show(self):
        print(f"{self.address.city},{self.address.pin}")

    
class Address:
    def __init__(self,city,pin):
        self.city = city
        self.pin = pin

a1 =Address("Pune",123)  
c1 = Customer("Amit",a1)   # Asscociation
c1.show()
```

```python

# composition(tight coupling)
class Engine:
    def start(self):
        print("Engine started")

class Car:
    def __init__(self):
        # Composition: Car creates Engine inside it
        self.engine = Engine()

    def drive(self):
        self.engine.start()
        print("Car is driving")

# Usage
c = Car()
c.drive()
```
